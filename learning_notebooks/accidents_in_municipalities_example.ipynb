```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
%matplotlib inline
```

```python
PROVIDER_AND_ID = "provider_and_id"
CITY = "accident_yishuv_name"
LAT = "latitude"
LONG = "longitude"
```

```python
csv_path = "involved_markers_hebrew.csv"
data = pd.read_csv(csv_path)
```

```python
def plot_city(data, city_name):
    city_data = data.loc[data[CITY] == city_name].drop_duplicates()
    fig, ax = plt.subplots(1, figsize=(7,7))
    ax.plot(city_data[LAT], city_data[LONG], "o")
    title_heb = "פיזור תאונות ב" + city_name
    ax.set_title(title_heb[::-1], fontsize=15)
    num_accidents = len(city_data)
    num_coords = len(city_data[[LAT, LONG]].drop_duplicates())
    txt_heb = str(num_accidents) + " תאונות ב-" + str(num_coords) + " מוקדים"
    fig.text(0.5, -.15, txt_heb[::-1], ha='center', transform=ax.transAxes,
            fontsize = 15)
    fig.show()
    
plot_city(data, "טבריה")
plot_city(data, "טייבה")
```
